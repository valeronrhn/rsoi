```python
from PIL import Image # Работа с изображением
from pylab import * # Отображение
import xmlrpc.client

# Режим отображения внутри ноутбука
%matplotlib inline
img = Image.open('11.bmp')
#img = Image.open('Jellyfish.jpg')

img_arr = array(img)

# Исходное изображение
gray()
imshow(img_arr)
```

```python
# Тест бинарной передачи данных
#pimg = img_arr.dumps()
pimg = pickle.dumps(img_arr) # универсально

img_bin = xmlrpc.client.Binary(pimg)

#img_bin2 = server.send_back_binary(img_bin)
img_bin2 = server.get_function('send_back_binary', img_bin)

#img_arr2 = np.loads(img_bin2.data)
img_arr2 = pickle.loads(img_bin2.data) # универсально

# Изображение после возрата с сервера
imshow(img_arr2)
```

```python

# Инверсия цвета изображения через сервер
def inv_color(img_arr_in):
    pimg = pickle.dumps(img_arr_in)
    img_bin = xmlrpc.client.Binary(pimg)
    
    #img_bin2 = server.color_inversion(img_bin)
    img_bin2 = server.get_function('color_inversion', img_bin)
    
    img_arr_out = pickle.loads(img_bin2.data)
    return img_arr_out

img_arr_inv = inv_color(img_arr)

# Изображение с инверсией цвета
imshow(img_arr_inv)
```

```python
#print (server.black_list_check(u'Петров'))
#print (server.black_list_check(u'Примеров'))
print( 'black_list_check:', server.get_function('black_list_check', 'Петров'))
```

```python
#Зеркальное отображение

pimg = pickle.dumps(img_arr)
img_bin = xmlrpc.client.Binary(pimg)
#mirr_img_bin = server.image_mirrorring(img_bin)
mirr_img_bin = server.get_function('image_mirrorring', img_bin)
img_arr = pickle.loads(mirr_img_bin.data)
imshow(img_arr)
```

```python
#Бинаризация изображения

pimg = pickle.dumps(img_arr)
img_bin = xmlrpc.client.Binary(pimg)
porog = 100
#binary_img_bin = server.binarisation_image(img_bin, porog)
binary_img_bin = server.get_function('binarisation_image', img_bin, porog)
binary_img_bin = pickle.loads(binary_img_bin.data)

imshow(binary_img_bin)
```
